# Data augmentation

## General 
Data augmentation aims to increase the diversity and size of the dataset, therefore, increasing the robustness and generalization of deep learning model. Data augmentation techniques often divided into traditional and deep learning based. Traditional techniques furtherly divided into geometric, photonic, noise and kernel+filter. Deep learning model often divided into GAN and style transfer. 

### Geometric
Geometric data augmentation techniques include: rotation (oftenly used, maybe it is best preserve information), however, rotation often create black region at the corner, can be solved using more complicated technique. 

Flipping is another technique, flip along horizontal and vertical axis or both, should be used with care based on our problem since it can change the semantic of the picture.

Cropping and resizing: randomly crop part of the given image and resizes the cropped part back to a certain size, this should be used carefully for image with multiple objects. Make sure the object being labeled is still inside the image.

Shifting and zooming

### Photometric
Colour juttering is a photometric data augmentation technique that employ either random colour manipulation or predetermined colour adjustments such as randomly chaning the brightness, contrast, blurring.

Noise transformation can be defined as photometric since it also changing the pixel value instead of the pizel location. Noise is random variation in brightness or colour information casued by enviromental condition of real-world data. Some popular types of noise is: Gaussian noise, salt and pepper and speckle noise.
- Gaussian noise: independent between pixel, independent of signal magnitude, additive noise.
- Salt and pepper: "spike noise" or "impulsive noise", it causes white and black pizels to appear at random.
- Speckle noise: multiplicative noise (dependent on the signal magnitude), generated by multiplying random values with different image pixels.
- Random erasing: Replace the values of pixel with random size regtangle. Careful with erasing the target prediction region.

### GAN

Adversial training or machine illusion attract considerable attention in deep learning community. By injecting a systematically created noise, CNN model outputs a completely different prediction, eventhough human eye cannot detect the difference. Synthetic images created by GANs, which even humans find difficult to distinguish from real image helps models significantly increase their robustness.


## Specific applciation

### Data augmentation in classification
Hussain et al. [68] evaluated various augmentation strategies on a **medical image dataset using VGG-16**. They demonstrated that the **flipping and Gaussian filter augmentation techniques yielded superior outcomes compared to the other ones.**

It was observed that CNN models trained from scratch benefited significantly from data augmentation, whereas pre-trained CNN models did not. In addition, it was discovered that **combinations of data augmentation techniques like rotation and varied illuminations could contribute most for CNN models trained from scratch in achieving excellent performance.**

In 2019, Zheng et al. [75] assessed the efficacy of neural style transfer using VGG16 on the Caltech 101 and Caltech 256 datasets, and the results demonstrated a two-percent gain in accuracy. Recent research has demonstrated that neural style transfer algorithms can apply the artistic style of one image to another image without altering the latterâ€™s high-level semantic content, showing that neural style transfer can be used for data augmentation to add more variation to the training dataset.

In 2020, Ismael et al. [76] employed data augmentation to solve the problem of insuffi-
cient training data and imbalanced classes in the MRI image classification task for brain
cancer. Various augmentation techniques, including horizontal and vertical flipping, rotating, shifting, zooming, shearing and brightness alteration, were utilised. They observed that each augmentation technique had different effects on the performance of distinct classes. For instance, manipulation of brightness yielded 96 percent accuracy for class one, whereas the rotation technique yielded 98 percent accuracy for the same class. For class two, these two techniques achieved a score of 99 percent with brightness and 98 percent with rotation. By combining all of the previously mentioned augmentation techniques, they were able to attain 99 percent overall accuracy, 4 percent improvement against the result obtained without data augmentation.
